here the output col diagnosis is inside the data set so we cant apply this technique, so remove that col diagnosis then store the remaining in any variable then apply encoding . after that add diagnosis in dataset back.

```python
print(breast['diagnosis'].unique())
```

```python
breast['diagnosis'] = breast['diagnosis'].map({'M': 1, 'B': 0})

breast

```

```python
#spliting data into test and train
```

```python
breast.shape
```

```python
x=breast.drop('diagnosis',axis=1)
y=breast['diagnosis']
```

```python
x.shape
```

```python
#train test split model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Feature scaling
```

```python
x_train
```

Here all the data is unscaled not a crt structure(not in crt range  formate) , all are flot so convert into structure scaling(change float number into crt range ).

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
```

```python
sc.fit(x_train)
```

```python
#after fitting, transform x_train into crt-range
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
# Train Model
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
y_pred=lg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
